```python

number = 10
print(number) #10

number -= 5
print(number) # 10 -5 = 5

number += 15
print(number) # 5 + 15 = 20

number *= 3
print(number) # 20 * 3 = 60

number /= 2
print(number) # 60 / 2 = 30.0

number %= 4
print(number) # 30.0 % 4 = 2.0
```

```python
#relational Operators
num1 = 10
num2 = 20

print(num1 == num2)
print(num1 > num2)
print(num1 < num2)
print(num1 >= num2)
print(num1 <= num2)
print(num1 != num2)
```

and

OR
0 or 0 = 0
0 or 1 = 1

1 or 1 = 1

```python
num1 = 25
num2 = 30
num3 = 40
num4 = 55
print("---------AND Operator---------")
print( num1 > num2 and num3 > num4 )
print( num1 > num2 and num3 < num4 )
print( num1 < num2 and num3 > num4 )
print( num1 < num2 and num3 < num4 )

print("---------OR Operator---------")
print( num1 > num2 or num3 > num4 )
print( num1 > num2 or num3 < num4 )
print( num1 < num2 or num3 > num4 )
print( num1 < num2 or num3 < num4 )

print("---------NOT Operator---------")
print(not(num1<num2))
```
