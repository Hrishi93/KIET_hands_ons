```python
student = {
    "name" : "Hrishikesh",
    "age" : 32,
    "address": {
        "city" : {
            "village" : "Panchgani",
            "district" : "Nagpur"
        },
        "state" : "Maharashtra",
        "pin_code": 445001
    }   
    }

print(student)
print(student["age"])
print(student["address"]["city"])

student["name"] = "Kishore"  #chnagle the value of name key
print(student)

student["address"]["city"]["village"] = "Hiwari"

print(student)

student["grade"]="HSC"

print(student)

student.pop("age")
print(student)
student.pop("address", "state")
print(student)
```
