```python
workshop1 = {"Alice", "Bob", "Charlie", "David"}
workshop2 = {"Eve", "Bob", "Frank", "David"}

print("Students who attended both workshops: ",workshop1.intersection(workshop2))
print("Students who attended workshops: ",workshop1.union(workshop2))
```
