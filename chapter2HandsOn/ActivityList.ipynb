```python
students_marks = {
    "Shyam" : 85,
    "Ram" : 92,
    "Mohan" : 78,
    "Sohan" : 68,
    "Geeta" : 50,
    "Krishna" : 72
}

for name , marks in students_marks.items():
    if marks >= 90:
        print(name , "has secured grade A")
    elif marks >= 80 and marks<= 89:
        print(name , "has secured grade B")
    elif marks >= 70 and marks<= 79:
        print(name , "has secured grade C")
    elif marks >= 60 and marks<= 69:
        print(name , "has secured grade D")
    else:
        print(name , "has secured grade F")
```
