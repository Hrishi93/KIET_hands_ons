```python
from sklearn.linear_model import LogisticRegression

x = [[1], [2], [3], [4], [5], [6], [7], [8]]
y = [0, 0, 0, 0, 1, 1, 1, 1]

model = LogisticRegression()
model.fit(x, y)


print("Predict for 4.5 class:", model.predict([[4.6]]))
print("Predict for 8 class:", model.predict([[8]]))
print("Predict for 8 class:", model.predict([[9]]))
print("Predict for 8 class:", model.predict([[-10]]))
```

```python
#manual Teast Third Part: SKLEARN
import numpy as np 

X = [[1], [2], [3], [4], [5], [6], [7], [8]]
Y = [0, 0, 0, 0, 1, 1, 1, 1]

#SIGmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#Logistic Regression function
def predict(X, weights=1.5 , bias=-6.7):
    z = weights * X + bias

    prob = sigmoid(z)
    return 1 if prob >= 0.5 else 0 
```
