#Age   #Income    #Purchess Capacity (I phone)
Young   HIGH         YES
Old     LOW          No
Young   LOW          NO
Old     HIGH         YES    

```python
import pandas as pd
data = {
    "age" : ["Young", "Young" , "Old" , "Old"],
    "income" : ["High", "Low", "High", "Low"],
    "Purchess_Capacity" : ["Yes", "No", "Yes", "No"]
}

df = pd.DataFrame(data)
```

```python
#Convert Data in numeric form

df['age'] = df['age'].map({'Young':0, 'Old':1})
df['income'] = df['income'].map({'Low':0, 'High':1})
df['Purchess_Capacity'] = df['Purchess_Capacity'].map({'No':0, 'Yes':1})
```

```python
#Selecting Model
from sklearn.tree import DecisionTreeClassifier
X = df[['age', 'income']]
y = df['Purchess_Capacity']
```

```python
#Putting data in Model
model_dtc = DecisionTreeClassifier(criterion='entropy')

#Fitting the data
model_dtc.fit(X, y)
```

```python
#prdicting the value 
#{'Young':0, 'Old':1})
#{'Low':0, 'High':1}
#{'No':0, 'Yes':1}
prediction = model_dtc.predict([[1, 0]])
print(prediction)
```

```python
#plot the tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(8,6))
tree.plot_tree(model_dtc, filled=True, feature_names=['age', 'income'], class_names=['No', 'Yes'])
plt.show()
```
