```python
#Read the flower image
img = cv2.imread('Images/flower.jpg')
```

```python
plt.imshow(img)
```

```python
#Annotation of Image
circle_Img = cv2.circle(img, (1000, 500), 200, (255, 0, 255), 30)
plt.imshow(circle_Img)

#want to add teaxt on image
text_img = cv2.putText(img, 
                       'Rose Flower', 
                       (300, 200), 
                       cv2.FONT_HERSHEY_SIMPLEX, 
                       10, 
                       (255, 255, 0), 
                       20)
```

```python
save_img = cv2.imwrite('text_img.jpg', text_img)
```

```python
#rotate the image
rotated_img = cv2.rotate(img, cv2.ROTATE_180)
plt.imshow(rotated_img)
```

```python
#Resize the image
resized_img = cv2.resize(img, (300, 200))
plt.imshow(resized_img)
```
