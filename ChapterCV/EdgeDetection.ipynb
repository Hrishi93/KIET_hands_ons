```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import Image, display
```

```python
img = cv2.imread('Couple.jpg')
```

```python
#Original Image
org = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(org)
```

```python
#Gray Scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#Soble Edge Detection
sobelx = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
sobely = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
sobelz = cv2.Sobel(src=gray, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
```

```python
#plotting the sobel edge detection
plt.figure(figsize=(15,10))
plt.subplot(141)
plt.imshow(sobelx, cmap='gray')
plt.subplot(142)
plt.imshow(sobely, cmap='gray')
plt.subplot(143)
plt.imshow(sobelz, cmap='gray')
```

```python
#Global Thresholding
retval, img_thresh_sobelx = cv2.threshold(sobelx, 50, 255, cv2.THRESH_BINARY)
retval, img_thresh_sobely = cv2.threshold(sobely, 50, 255, cv2.THRESH_BINARY)
retval, img_thresh_sobelz = cv2.threshold(sobelz, 50, 255, cv2.THRESH_BINARY)
```

```python
#showing the thresholded images
plt.figure(figsize=(15,10))
plt.subplot(141)
plt.imshow(img_thresh_sobelx, cmap='gray')
plt.subplot(142)
plt.imshow(img_thresh_sobely, cmap='gray')
plt.subplot(143)
plt.imshow(img_thresh_sobelz, cmap='gray')
plt.show()
```
