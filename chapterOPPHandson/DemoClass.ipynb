```python
#WAP for Student Info using class

class Student:
    roll_no = 1201
    fname = "Hrishikesh"
    lname = "Mahure"
    city = "Nagpur"
    branch = "CSE"
    cgpa = 7.8

    #user define method
    def display_genral_info(self):
        print("Name :" , self.fname , self.lname)
        print("city :", self.city)

    def display_academic_info(self):
        print("Roll No " , self.roll_no)
        print("Branch :" , self.branch)
        print("CGPA :", self.cgpa)
    
```

```python
#Instance of class
std = Student()
print("------Academic------")
std.display_academic_info()
print("------Genral------")
std.display_genral_info()

#print(std.roll_no)
#print(std.name)
#print(std.branch)
```

```python
class Employee:

    salary = 75000 #Global Variable

    def __init__(self, new_name , new_age, new_city):
        self.name=new_name
        self.age=new_age
        self.city=new_city
        #print(self.salary)
        #print(self.dept)  #dept is a Local Varibale inside employee_info()

    def employee_info(self , dept):
        print("--------New Employee-----")
        print("Name ", self.name)
        print("Age ", self.age)
        print("city :", self.city)
        print("Salary :", self.salary)
        print("Department :" , dept)

    def good_bye(self):
        return "Bye Bye"  
        

#take the at the time of Object creation and diplay using employee_info()
```

```python
emp = Employee("Ramesh" , 32 , "Nagpur")
emp.employee_info("HR")
print(emp.good_bye())

emp = Employee("Suresh", 36 , "Kakinada")
emp.employee_info("FINANCE")
```
