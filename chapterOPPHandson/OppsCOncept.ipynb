```python
class Parent:
    saving = 1000000
    house = "3BHK"
    shares = 300000

    def loan(self):
        print("Have 20 Lakh home Loan")
```

```python
class Child(Parent):
    salary = 25000
    def property(self):
        print("1BHK at Hyderabd")

    #Overiding
    def loan(self):
        print("Loan amount is zero by child")
```

```python
c1 = Child() #instance ob Child
print(c1.salary)
c1.property()

print(c1.saving)
c1.loan()

c1.saving += 50000 
print(c1.saving)
```

```python
#multiple Inheritance 
#multiple parent class 
#one child class

class Parent_one:
    saving = 250000
    def display(self):
        print("Method of Parent one")
```

```python
class Parent_two:
    property = 500000

    def show(self):
        print("Metho of Parent Two")
```

```python
class Child(Parent_one , Parent_two):
    def __init__(self):
        print("In a Child")
```

```python
ch1 = Child()
print(ch1.saving) #Parent One Attribute
print(ch1.property) #Prent two attribute
ch1.display()
ch1.show()
```

```python
#Encapsulation
#wraping a data (Variable) in a Method
class Employee:
    name = "Hrishkesh"
    age = 32
    city = "Nagpur"

    def __init__(self):
        print("----Employee Details------")

    #getter
    def get_name(self):
        return self.name
    
    def get_age(self):
        return self.age
    
    def get_city(self):
        return self.city
    
    #setter
    def set_name(self , new_name):
        self.name = new_name

    def set_age(self , new_age):
        self.age = new_age

    def set_city(self, new_city):
        self.city = new_city     
```

```python
emp = Employee()
#print(emp.name)
print(emp.get_name())
print(emp.get_age())
print(emp.get_city())

print("---------Set Employee----------")
emp.set_name("Ramesh")
emp.set_age(35)
emp.set_city("Kakinada")
print(emp.get_name())
print(emp.get_age())
print(emp.get_city())
```

```python
emp = Employee()

emp.name = "Ramesh"
emp.age = 35
emp.city = "Kakinada"
print(emp.name)
print(emp.age)
print(emp.city)
```

```python
#Same inside same class #overloading
class Addition:
    num1 = 20
    num2 = 30

    def add(self):
        print(self.num1+self.num2)

    def add(self, num1):
        self.name1 = num1
        print(self.num1 + self.num2)

    def add(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        print(self.num1 + self.num2)
```

```python
a1 = Addition()
a1.add(10, 20)
```
