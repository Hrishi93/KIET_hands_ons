```python
class Vehicle:
    def __init__(self , make , model , year):
        self.make = make
        self.model = model
        self.year = year

    def vehicle_details(self):
        print("Make :", self.make)
        print("Model :", self.model)
        print("Year :", self.year)

    def update_model(self, model):
        self.model=model
```

```python
obj_veh = Vehicle("Tata" , "Nexon", 2024)

print("-----------before Update-----------------")
obj_veh.vehicle_details()

print("------------after Update------------")
obj_veh.update_model("Altroz")
obj_veh.vehicle_details()
```
